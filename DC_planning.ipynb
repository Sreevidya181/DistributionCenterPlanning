```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
# df = pd.read_csv('DC_Planning.csv')
df = pd.read_csv('/content/drive/MyDrive/Target Upcurve/Project/DC_Planning.csv')

# Segment guests based on categorical preferences
segmented_guests = df.groupby(['guest_segmentation', 'shipping_preferences']).size().reset_index(name='Count')

# Plot segmented guests
plt.figure(figsize=(12, 8))
sns.barplot(data=segmented_guests, x='guest_segmentation', y='Count', hue='shipping_preferences')
plt.title('Segmented Guests by Guest Segmentation and Shipping Preferences')
plt.xlabel('Guest Segmentation')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()

# Analyze the distribution of order-related metrics within each segment
numerical_columns = ['order_frequency', 'returns_rate', 'avg_order_value']

for segment in df['guest_segmentation'].unique():
    plt.figure(figsize=(12, 8))
    segment_df = df[df['guest_segmentation'] == segment]
    for column in numerical_columns:
        sns.histplot(data=segment_df, x=column, kde=True, label=column, alpha=0.5)
    plt.title(f'Distribution of Order Metrics for {segment} Guests')
    plt.xlabel('Order Metrics')
    plt.ylabel('Frequency')
    plt.legend()
    plt.show()
```
